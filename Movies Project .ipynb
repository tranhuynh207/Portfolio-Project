```python
## This project is to find out correlations of different factors affecting movie revenue
## Results show that gross revenues have correlations with the no. of votes and the movie budgets


#Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#Read data
df = pd.read_csv ('/Users/Admin/Downloads/movies.csv')
```

```python
#Look at the data 
df.head()
```

```python
# Check missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Drop missing data
df = df.dropna()
```

```python
# Check data type
print (df.dtypes)
```

```python
# Change data type for columns 
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
df['votes']= df['votes'].astype('int64')
df.head ()
```

```python
# Any outliers?

df.boxplot(column=['gross'])
```

```python
# Order Data

df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
# Check duplicate in columns
df['company'].sort_values (ascending= False)
```

```python
# Drop duplicate
df.drop_duplicates()
```

```python
# Plot 'gross' vs 'budget' using seaborn
sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color":"red"}, line_kws={"color":"black"})
```

```python
# Seem correlations between 'gross' and 'budget'
```

```python
# Correlation
df.corr(method='pearson') #'spearman', 'Kendall'
```

```python
# High correlation between 'budget' and 'gross'
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Other 'string' variables such as 'Company', 'Country' might have correlations too?
# Create a dataframe changing these variables as categories type

df_numberized = df
for col_name in df_numberized.columns :
    if (df_numberized[col_name].dtype == 'object'):
        df_numberized[col_name] = df_numberized[col_name].astype('category')
        df_numberized[col_name] = df_numberized[col_name].cat.codes
        
df_numberized.head ()
```

```python
# Correlations within the numberized df presented with heatmap

correlation_matrix = df_numberized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Correlations within the numberized df

df_numberized.corr(method='pearson')
```

```python
# Let's pull out pairs with correlation coef > 0.5
correlation_matrix = df_numberized.corr(method='pearson')
corr_pairs = correlation_matrix.unstack()

sorted_pairs = corr_pairs.sort_values()

positive_corr = sorted_pairs [(sorted_pairs)>0.5]

positive_corr.drop_duplicates() #easier to see by dropping duplicates
```

```python
# 'Gross' has correlations with 'votes' and 'budget'
```

```python
# How about negative correlation? 
negative_corr = sorted_pairs [(sorted_pairs)<-0.5]
negative_corr
```
